```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

import seaborn as sns
sns.set(style="whitegrid")
```

```python
results = pd.read_csv('../data/Model_Results.csv')
```

```python
results
```

```python
results.loc[results.Model == 'Classifier Chain', 'Model'] = 'Classifier Chain\n LinearSVC'
```

```python
g = sns.catplot(x="Model", y="Hamming_Loss", hue="Upsampled", data=results,
                height=6, kind="bar", palette="muted")

for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(30)
g.despine(left=True)
g.set_ylabels("Hamming Loss")
g.set(ylim=(0.02, 0.027))
g.fig.set_size_inches(12,6)
g.fig.suptitle('Comparison of Hamming Loss Scores', fontweight = 'bold', fontsize = 18, y = 1.05);

g.savefig('../images/hamming.png')
```

```python
g = sns.catplot(x="Model", y="MicroF1", hue="Upsampled", data=results,
                height=6, kind="bar", palette="muted")

for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(30)
g.despine(left=True)
g.set_ylabels("MicroF1")
g.set(ylim=(0.66, 0.71))
g.fig.set_size_inches(12,6)
g.fig.suptitle('Comparison of Micro F1 Scores', fontweight = 'bold', fontsize = 18, y = 1.05);
g.savefig('../images/microf1.png')
```

```python
g = sns.catplot(x="Model", y="MacroF1", hue="Upsampled", data=results,
                height=6, kind="bar", palette="muted")

for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(30)
    for label in ax.get_yticklabels():
        label.set_fontsize(14)
g.despine(left=True)
g.set_ylabels("Macro F1 Score", fontsize = 14)
g.set(ylim=(0.57, 0.66))
g.fig.set_size_inches(12,6)
g.fig.suptitle('Comparison of Macro F1 Scores', fontweight = 'bold', fontsize = 18, y = 1.05);
g.savefig('../images/macrof1.png')
```
